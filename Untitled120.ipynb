```python
#!/usr/bin/env python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
from collections import Counter

# Path to Hamlet text and mask image
hamlet_path = r"C:\Users\SRIMAN\Downloads\hamlet\Hamlet.txt"
mask_path = r"C:\Users\SRIMAN\Downloads\mask oval ..png"  # Updated mask path

# Read the text
with open(hamlet_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Remove punctuation and convert to lower case
text = text.lower()

# Create a stopword set
stopwords = set(STOPWORDS)

# Load the mask image
mask = np.array(Image.open(mask_path))

# Ensure that the mask is binary (black and white) for clear word shaping
mask = np.where(mask > 128, 255, 0)

# Generate the word cloud
wordcloud = WordCloud(background_color='white', max_words=2000, mask=mask,
                      stopwords=stopwords, contour_width=3, contour_color='steelblue',
                      colormap='cool').generate(text)

# Display the word cloud inside the oval
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Create a frequency count of words
word_counts = Counter(text.split())
top_20_words = word_counts.most_common(20)

# Unzip the words and counts for plotting
words, counts = zip(*top_20_words)

# Generate a bar chart for the top 20 words
plt.figure(figsize=(10, 6))
plt.barh(words, counts, color='steelblue')
plt.xlabel('Frequency')
plt.title('Top 20 Words in Hamlet')
plt.gca().invert_yaxis()  # To display the highest count at the top
plt.show()
```
